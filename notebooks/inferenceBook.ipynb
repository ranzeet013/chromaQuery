```python
from langchain.vectorstores import Chroma
from langchain_groq import ChatGroq
from langchain.chains.question_answering import load_qa_chain
from langchain.embeddings import SentenceTransformerEmbeddings

persist_directory = "/content/drive/MyDrive/chroma_db"

embeddings = SentenceTransformerEmbeddings(model_name="all-MiniLM-L6-v2")
new_db = Chroma(persist_directory=persist_directory, embedding_function=embeddings)

llm = ChatGroq(
    temperature=0,
    groq_api_key='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',
    model_name="llama3-8b-8192"
)

chain = load_qa_chain(llm, chain_type="stuff", verbose=True)

def query_chroma(query, db):
    """Retrieve relevant documents and generate an answer using LLM."""
    matching_docs = db.similarity_search(query, k=2)
    answer = chain.run(input_documents=matching_docs, question=query)
    return answer

query = "Explain Retrieval-Augmented Generation?"
answer = query_chroma(query, new_db)

print("\n📝 **Answer:**")
print(answer)
```
